Rectified Signal

```python
import matplotlib.pyplot as plt
signal=abs(signal)

plt.xlabel('Number of samples')
plt.ylabel('Amplitude')
plt.title('EMG')
plt.plot(signal)
```

```python
#frequency domain
from scipy.fftpack import fft
import numpy as np

Freq=1000
N = len(signal)
n = np.arange(N)
T = N/Freq
freq = n/T
X = fft(signal)
plt.figure(figsize = (12, 6))
plt.subplot(121)

plt.stem(freq, np.abs(X), 'b', \
         markerfmt=" ", basefmt="-b")
plt.xlabel('Freq (Hz)')
plt.ylabel('FFT Amplitude |X(freq)|')
plt.xlim(20, 450)  # useful range of frequencies in EMG
```
