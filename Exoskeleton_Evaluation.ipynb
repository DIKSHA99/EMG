Evaluation of Muscle amplitudes for exoskeleton on and off state for Tibialis Anterior and Gastrocnemius Medialis

```python
import pandas as pd
import numpy as np
data=pd.read_excel('EXO-ON-OFF.xlsx','Sheet1')
data1=pd.read_excel('EXO-ON-OFF.xlsx','Sheet2')
data4=pd.read_excel('EXO-ON-OFF.xlsx','Sheet5')

#removing nanas
data=data.replace([np.inf, -np.inf], np.nan, inplace=False)
data=pd.DataFrame(data)
print(np.shape(data))

data1=data1.replace([np.inf, -np.inf], np.nan, inplace=False)
data1=pd.DataFrame(data1)
print(np.shape(data1))


data4=data4.replace([np.inf, -np.inf], np.nan, inplace=False)
data4=pd.DataFrame(data4)
print(np.shape(data4))

#assigning columns for particular sheets
Tibialis_exo_on = data.iloc[1:299061,0].values
Tibialis_exo_off = data.iloc[1:299061,1].values
Gastrocnemius_exo_on = data1.iloc[1:299061,0].values
Gastrocnemius_exo_off = data1.iloc[1:299061,1].values
Spinae_exo_on = data4.iloc[1:299061,0].values
Spinae_exo_off = data4.iloc[1:299061,1].values
```

```python
import matplotlib.pyplot as plt
plt.subplot(2,1,1)
plt.xlabel('Number of samples')
plt.ylabel('Amplitude')
plt.title('TIBIALIS')
plt.plot( abs(Tibialis_exo_on), 'b-')
plt.plot(abs(Tibialis_exo_off), 'y-', linewidth=1)
plt.legend(["Tibialis_exo_on", "Tibialis_exo_off"], loc ="upper left")


```

```python
plt.subplot(2,1,1)
plt.xlabel('Number of samples')
plt.ylabel('Amplitude')
plt.title('GASTROCNEMIUS')
plt.plot( abs(Gastrocnemius_exo_on), 'b-')
plt.plot(abs(Gastrocnemius_exo_off), 'y-', linewidth=1)
plt.legend(["Gastrocnemius_exo_on", "Gastrocnemius_exo_on"], loc ="upper left")
```
